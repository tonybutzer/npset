````python
def plot_temp(temp,cmap):
    fig, ax = plt.subplots(1, 1, figsize=(10, 8))
    plt.gca().invert_yaxis()
    viz_tools.set_aspect(ax)
    mesh = ax.pcolormesh(temp,cmap=cmap)
    fig.colorbar(mesh)
    dest = 'temperature'+cmap+'.png'
    plt.savefig(dest)
    plt.cla()
````

````python
for cmap in list:
    pcmap = '<'+cmap+'>'
    print(pcmap)
    cmap = cmap.replace(' ','')
    pcmap = '<'+cmap+'>'
    print(pcmap)
    plot_temp(temp,cmap)
````
