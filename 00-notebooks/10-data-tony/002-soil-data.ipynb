```python
# Intercept_us_filled.tif
from matplotlib import pyplot
tif_file = '/vsis3/ga-et-data/inputsv0/soilData/Intercept_us_filled.tif'
with rasterio.open(tif_file) as src:
    myThumb = src.read(1, out_shape=(1, int(src.height // 8), int(src.width // 8)))

print(myThumb.shape)
pyplot.imshow(myThumb, cmap='prism')

pyplot.show()
```

```python
from rasterio.plot import show_hist
show_hist(myThumb, bins=50, lw=0.0, stacked=False, alpha=0.3, histtype='stepfilled', title="Histogram")
```

```python
array = myThumb
array[(array < 0)] = 0
```

```python
show_hist(array, bins=50, lw=0.0, stacked=False, alpha=0.3, histtype='stepfilled', title="Histogram")
```
