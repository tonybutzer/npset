```python

import os
 
path = './data'
file_list=[]
files = os.listdir(path)
for name in files:
    if '.png' in name:
        print(name)
        name = './data/'+name
        print(name)
        file_list.append(name)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
%matplotlib inline
#%matplotlib notebook

#fig = plt.figure()
fig, ax = plt.subplots(1, 1, figsize=(8, 8))

ims=[]
for file in file_list:
    print(file)
    img = mgimg.imread(file)
    im = plt.imshow(img, animated=True)
    ims.append([im])

ani = animation.ArtistAnimation(fig, ims, interval=500, blit=True,
                                repeat_delay=1000)

ani.save('10hot_temp.gif')
```

```python
from matplotlib import pyplot as plt  
from matplotlib import animation  
import matplotlib.image as mgimg
import numpy as np

#set up the figure
fig = plt.figure()
ax = plt.gca()

#print(file_list)

#initialization of animation, plot array of zeros 
def init():
    imobj.set_data(np.zeros((100, 100)))

    return  imobj,

def aanimate(i):
    ## Read in picture
    fname = file_list[i]
    print("FNAME=",fname)

    ## here I use [-1::-1], to invert the array
    # IOtherwise it plots up-side down
    img = mgimg.imread(fname)[-1::-1]
    imobj.set_data(img)

    return  imobj,


## create an AxesImage object
imobj = ax.imshow( np.zeros((100, 100)), origin='lower', alpha=1.0, zorder=1, aspect=1 )


anim = animation.FuncAnimation(fig, aanimate, init_func=init, repeat = True,
                               frames=range(1,100), interval=1200, blit=True, repeat_delay=1000)

plt.show()
```
