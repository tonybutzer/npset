```python
import numpy as np
import xarray as xr
```

```python
from os import listdir
from os.path import isfile, join
mypath='/mnt/ga-et-data/Cloud_Veg_ET/Data/ETO'
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
```

```python
full_filename = mypath +'/' + onlyfiles[0]

ary = np.load(full_filename)
baseline_shape = ary.shape
print(baseline_shape)

datasets = []

for file in onlyfiles:
    full_filename = mypath +'/' + file
    print(full_filename)
    ary = np.load(full_filename)
    ary = np.flip(ary, axis=0)
    ary[(ary < 0)] = np.nan

    if (ary.shape == baseline_shape):
        print('YAY!')
    else:
        print('BOO! BOO!')
    da = xr.DataArray(ary,dims=['row', 'col'])
    ds = da.to_dataset(name=file)
    datasets.append(ds)

DS = xr.merge(datasets)
    
```
