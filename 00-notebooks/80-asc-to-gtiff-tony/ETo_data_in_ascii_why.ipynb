```python
f = open('ETo.asc', 'r')
```

```python
data = f.read()
```

```python
#data
```

```python
lines = data.split('\n')
```

```python
lines[0]
```

```python
for i in range(6):
    print(i, lines[i])
```

```python
lines[6]
```

```python
import numpy as np
a = np.zeros((560,510))
```

```python
#dir(np)
```

```python
a.shape
```

```python
for row in range(560):
    vals = lines[row+6].split(' ')
    for col in range(510):
        # print(row,col)
        a[row,col]=vals[col]
        
```

```python
a
```

```python
import matplotlib.pyplot as plt
plt.imshow(a)
```

```python
a[(a == -9999)]=np.nan
```

```python
plt.imshow(a)
```

```python
outfile = 'sane_numpy_ETo.npy'
```

```python
np.save(outfile, a)
```
